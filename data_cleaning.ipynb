## Data Cleaning

```python
# imports
import pandas as pd
```

```python
# reading the data
data_2022 = pd.read_csv("data_22.csv")
data_2023 = pd.read_csv("data_23.csv")
data_2024 = pd.read_csv("data_24.csv")
```

### 2022 Data

```python
# look at number of rows and columns
data_2022.shape
```

```python
# list all columns
data_2022.columns.tolist()
```

```python
# keep only required columns
data_2022 = data_2022[[
    "uri",
    "track_name",
    "artist_names",
    "peak_rank",
    "weeks_on_chart",
    "danceability",
    "energy",
    "key",
    "loudness",
    "mode",
    "speechiness",
    "acousticness",
    "tempo"
]]
```

```python
# look at the dataset high-level
data_2022.head()
```

```python
# check data types and if null values are allowed
data_2022.info()
```

```python
# null values are not allowed but to check
data_2022.isnull().sum()
```

#### standardising 2022 data types

```python
# Clean text columns
text_cols_2022 = ["uri", "artist_names", "track_name"]

for col in text_cols_2022:
    data_2022[col] = data_2022[col].astype("string")

# Integer columns
int_cols_2022 = ["peak_rank", "weeks_on_chart", "key", "mode",]

for col in int_cols_2022:
    data_2022[col] = pd.to_numeric(data_2022[col], errors="coerce").astype("Int64")

# Float columns
float_cols_2022 = ["danceability", "energy", "loudness", "speechiness", "acousticness", "tempo"]

for col in float_cols_2022:
    data_2022[col] = pd.to_numeric(data_2022[col], errors="coerce")

# give key numbers their actual key values
key_map = {
0:"C",1:"C#",2:"D",3:"D#",4:"E",
5:"F",6:"F#",7:"G",8:"G#",9:"A",10:"A#",11:"B"
}

data_2022["key"] = data_2022["key"].map(key_map)

# same for Major or Minor
mode_map = {1:"Major",0:"Minor"}
data_2022["mode"] = data_2022["mode"].map(mode_map)

# convert them into string
data_2022["key"] = data_2022["key"].astype("string")
data_2022["mode"] = data_2022["mode"].astype("string")

# Check result
data_2022.info()
```

### 2023 Data

```python
# repeat the same process for 2023 dataset
data_2023.shape
```

```python
# check columns
data_2023.columns.tolist()
```

```python
# remove columns which are not required, keep only required
data_2023 = data_2023[[
    "track_name",
    "artist(s)_name",
    "streams",
    "bpm",
    "key",
    "mode",
    "danceability_%",
    "valence_%",
    "energy_%",
    "acousticness_%",
    "instrumentalness_%",
    "liveness_%",
    "speechiness_%"
]]
```

```python
data_2023.head()
```

```python
data_2023.info()
```

```python
data_2023.isnull().sum()
```

#### standardize 2023 data types

```python
# Rename artist column to match other datasets
data_2023 = data_2023.rename(columns={
    "artist(s)_name": "artist_names"
})

# Clean text columns
text_cols_2023 = ["track_name", "artist_names", "key", "mode"]

for col in text_cols_2023:
    data_2023[col] = data_2023[col].astype("string")

# Clean and convert streams safely
data_2023["streams"] = (
    data_2023["streams"]
    .astype("string")
    .str.replace(",", "", regex=False)
    .str.strip()
)

data_2023["streams"] = pd.to_numeric(data_2023["streams"], errors="coerce")

# Drop rows where streams is invalid/missing
data_2023 = data_2023.dropna(subset=["streams"])

# Now convert to integer
data_2023["streams"] = data_2023["streams"].astype("int64")

# Numeric columns
numeric_cols_2023 = [
    "bpm",
    "danceability_%",
    "valence_%",
    "energy_%",
    "acousticness_%",
    "instrumentalness_%",
    "liveness_%",
    "speechiness_%"
]

for col in numeric_cols_2023:
    data_2023[col] = pd.to_numeric(data_2023[col], errors="coerce")

# Convert percentage columns to decimals to match 2022 dataset
cols_pct = [
    "danceability_%",
    "valence_%",
    "energy_%",
    "acousticness_%",
    "instrumentalness_%",
    "liveness_%",
    "speechiness_%"
]

data_2023[cols_pct] = data_2023[cols_pct] / 100

# Rename columns so they match 2022 naming
data_2023 = data_2023.rename(columns={
    "danceability_%": "danceability",
    "valence_%": "valence",
    "energy_%": "energy",
    "acousticness_%": "acousticness",
    "instrumentalness_%": "instrumentalness",
    "liveness_%": "liveness",
    "speechiness_%": "speechiness"
})

# Keep key and mode as cleaned text
data_2023["key"] = data_2023["key"].replace("nan", pd.NA)
data_2023["mode"] = data_2023["mode"].replace("nan", pd.NA)

# Check result
data_2023.info()
```

#### Fix the 2023 artist column name so it matches too


```python
data_2023 = data_2023.rename(columns={
    "artist(s)_name": "artist_names"
})

data_2023.columns
```

#### Standardize column names in the 2024 dataset

```python
data_2024.columns = (
    data_2024.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

data_2024.columns
```

#### Rename the key identity columns in 2024 to match the other datasets

```python
data_2024 = data_2024.rename(columns={
    "track": "track_name",
    "artist": "artist_names"
})

data_2024.columns
```

```python
data_2024.dtypes
```

#### Convert spotify_streams and the other major numeric columns to a real numeric column

```python
# Clean text columns
text_cols_2024 = ["track_name", "artist_names", "isrc"]

for col in text_cols_2024:
    data_2024[col] = data_2024[col].astype("string")

# Clean and convert spotify_streams safely
data_2024["spotify_streams"] = (
    data_2024["spotify_streams"]
    .astype("string")
    .str.replace(",", "", regex=False)
    .str.strip()
)

data_2024["spotify_streams"] = pd.to_numeric(data_2024["spotify_streams"], errors="coerce")

# Drop rows where spotify_streams is invalid/missing
data_2024 = data_2024.dropna(subset=["spotify_streams"])

# Convert spotify_streams to integer
data_2024["spotify_streams"] = data_2024["spotify_streams"].astype("int64")

# Convert remaining numeric columns
data_2024["all_time_rank"] = pd.to_numeric(data_2024["all_time_rank"], errors="coerce").astype("Int64")
data_2024["track_score"] = pd.to_numeric(data_2024["track_score"], errors="coerce")

# Check result
data_2024.info()
```

#### Standardize text formatting for song and artist names
- remove leading/trailing spaces
- convert to lowercase
- remove double spaces

```python
def clean_text(series):
    return (
        series.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(r"\s+", " ", regex=True)
    )

data_2022["track_name"] = clean_text(data_2022["track_name"])
data_2022["artist_names"] = clean_text(data_2022["artist_names"])

data_2023["track_name"] = clean_text(data_2023["track_name"])
data_2023["artist_names"] = clean_text(data_2023["artist_names"])

data_2024["track_name"] = clean_text(data_2024["track_name"])
data_2024["artist_names"] = clean_text(data_2024["artist_names"])
```

Dropping rows based on all_time_rank would remove a large portion of the dataset (over 70%), which would significantly reduce the dataset size and bias the analysis. Therefore, all_time_rank is kept as NULL where missing.
